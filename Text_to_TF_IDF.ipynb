```python
data = pd.DataFrame({'text':sentences})
```

```python
data.head()
```

```python
# Now lets convert the text into numbers using TF-IDF

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(data['text'])
```

```python
# Now the data is in SParse-Format, we have to bring it to normal 2D array

tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns = tfidf.get_feature_names_out())
tfidf_df.head()
```

```python
tfidf_df['tfidf_score'] = tfidf_df.mean(axis=1)

# Keep only the single column
result = pd.DataFrame({'sentence': sentences, 'tfidf_score': tfidf_df['tfidf_score']})
```
